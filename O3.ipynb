```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
import time
import pandas as pd  # Thêm pandas để tạo DataFrame

driver = webdriver.Chrome() 
driver.get("https://air.plumelabs.com/air-quality-in-Da%20Nang-6E4g")  # Thay thế bằng URL chứa phần tử

# Nhấp vào nút "Year" để chuyển đến dữ liệu theo năm
year_button = driver.find_element(By.XPATH, "//li[@data-time-frame='year']")
year_button.click()
time.sleep(2)  # Đợi trang load dữ liệu sau khi nhấp

# Cuộn trang xuống 100 pixel
driver.execute_script("window.scrollBy(0, 200);")
time.sleep(1)

# Tìm phần tử ô tròn bằng id
circle_element = driver.find_element(By.ID, "chart__point")
actions = ActionChains(driver)

# Click and hold phần tử ô tròn
actions.click_and_hold(circle_element).perform()

# List để lưu trữ dữ liệu ngày và chỉ số O3
data = []

# Set để lưu trữ các ngày đã xuất hiện
unique_dates = set()

# Vòng lặp di chuyển chuột và lấy thông tin ngày và chỉ số O3
for _ in range(485):  # Thay đổi số lần lặp tùy ý
    actions.move_by_offset(-1, 0).perform()

    # Lấy ngày hiện tại từ phần tử ngày tháng
    date_element = driver.find_element(By.CSS_SELECTOR, "p#chart-date.report__lead")
    current_date = date_element.text

    # Chỉ xử lý nếu ngày hiện tại chưa xuất hiện trước đó
    if current_date not in unique_dates:
        unique_dates.add(current_date)

        # Lấy chỉ số O3
        o3_element = driver.find_element(By.XPATH, "/html/body/div[2]/div[2]/div[5]/div[2]/ul/li[4]/div[1]")
        o3_value = o3_element.text
        time.sleep(2)

        # Thêm dữ liệu vào list
        data.append([current_date, o3_value])

        # In ra màn hình ngày và chỉ số O3
        print(f"Ngày: {current_date}, O3: {o3_value}")

# Chuyển list thành DataFrame của pandas
df = pd.DataFrame(data, columns=["Day", "O3"])

# Xuất DataFrame thành file CSV
df.to_csv("o3_data_pandas.csv", index=False)

# Thả chuột (release)
actions.release().perform()

# Đóng trình duyệt
driver.quit()
```
